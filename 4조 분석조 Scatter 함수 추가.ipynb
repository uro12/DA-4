연도별  인구1인당전력소비량 (kWh/년) 변화율 

```python
import csv
import matplotlib.pyplot as plt

# 파일 열기
f = open('연도별_전력수급_실적.csv', encoding='cp949')
data = csv.reader(f)
next(data)  # 헤더를 건너뛰기

# 시점과 '인구1인당전력소비량 (kWh/년)' 추출
years = []
power_per_capita = []

for row in data:
    years.append(row[0])  # 시점(연도)
    power_per_capita.append(float(row[-1]))  # '인구1인당전력소비량 (kWh/년)' (마지막 열)

# 파일 닫기
f.close()

# 그래프 설정
plt.plot(years, power_per_capita, marker='o', linestyle='-', color='b')  # 시점 별 전력 소비량 그래프
plt.title('인구1인당전력소비량 (kWh/년) 시점별 변화')
plt.xlabel('연도')
plt.ylabel('인구 1인당 전력 소비량 (kWh/년)')
plt.xticks(rotation=45)  # 연도 레이블 회전
plt.grid(True)

# 그래프 출력
plt.tight_layout()  # 레이아웃 조정
plt.show()
```

연도별 인구1인당전력소비량 (kWh/년)과 경제성장률 (%) 비교

```python
import csv
import matplotlib.pyplot as plt

# '연도별_전력수급_실적.csv' 파일을 열고 데이터 읽기
f = open('연도별_전력수급_실적.csv', encoding='cp949')
data = csv.reader(f)
next(data)  # 헤더를 건너뛰기

# 연도, '인구1인당전력소비량 (kWh/년)', '경제성장률 (%)' 데이터 추출
years = []
power_per_capita = []
growth_rate = []

for row in data:
    years.append(row[0])  # 시점(연도)
    power_per_capita.append(float(row[-1]))  # '인구1인당전력소비량 (kWh/년)' (마지막 열)
    growth_rate.append(float(row[1]))  # '경제성장률 (%)' (두 번째 열)

f.close()  # 파일 닫기

# 시각화를 위한 두 개의 서브플롯 생성
fig, ax1 = plt.subplots(figsize=(10, 6))

# 첫 번째 그래프: '인구1인당전력소비량 (kWh/년)'
ax1.plot(years, power_per_capita, marker='o', color='b', label='인구1인당전력소비량 (kWh/년)')
ax1.set_xlabel('시점 (연도)')
ax1.set_ylabel('1인당 전력 소비량 (kWh/년)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# x축의 tick 위치를 설정 (set_ticks 호출)
ax1.set_xticks(range(len(years)))  # x축 위치 설정
ax1.set_xticklabels(years, rotation=45)  # x축 레이블 설정

# 두 번째 y축을 사용하여 '경제성장률 (%)' 그래프 추가
ax2 = ax1.twinx()
ax2.plot(years, growth_rate, marker='s', color='g', linestyle='--', label='경제성장률 (%)')
ax2.set_ylabel('경제성장률 (%)', color='g')
ax2.tick_params(axis='y', labelcolor='g')

# 타이틀 추가
plt.title('연도별 1인당 전력 소비량 (kWh/년)과 경제성장률 (%) 비교')

# 그래프 출력
fig.tight_layout()  # 레이아웃 조정
plt.grid(True)
plt.show()
```

연도별 1인당 전력 소비량과 1인당 GDP 비교

```python
import csv
import matplotlib.pyplot as plt

# '1인당_국내총생산__당해년가격.csv' 파일을 열고 데이터 읽기
gdp_file = '1인당_국내총생산__당해년가격2.csv'
power_file = '연도별_전력수급_실적.csv'

# GDP 데이터 읽기
gdp_data = {}
f = open(gdp_file, encoding='cp949')  # 파일 열기
data = csv.reader(f)
header = next(data)  # 첫 번째 줄(헤더) 건너뛰기
for row in data:
    if row[0] == '대한민국':  # 대한민국 데이터만 사용
        for year, value in zip(header[2:], row[2:]):  # GDP 값이 세 번째 열부터 시작
            gdp_data[year] = float(value)
f.close()  # 파일 닫기

# 전력 소비량 데이터 읽기
power_data = {}
f = open(power_file, encoding='cp949')  # 파일 열기
data = csv.reader(f)
header = next(data)  # 첫 번째 줄(헤더) 건너뛰기
for row in data:
    year = row[0]
    if year in gdp_data:  # GDP 데이터에 있는 연도만 포함
        power_data[year] = float(row[-1])  # '인구1인당전력소비량 (kWh/년)' 열 사용
f.close()  # 파일 닫기

# 공통 연도 추출
years = sorted(set(gdp_data.keys()) & set(power_data.keys()))
gdp_values = [gdp_data[year] for year in years]
power_values = [power_data[year] for year in years]

# 시각화를 위한 두 개의 서브플롯 생성
fig, ax1 = plt.subplots(figsize=(10, 6))

# 첫 번째 그래프: 1인당 전력 소비량
ax1.plot(years, power_values, marker='o', color='b', label='1인당 전력 소비량 (kWh/년)')
ax1.set_xlabel('연도')
ax1.set_ylabel('1인당 전력 소비량 (kWh/년)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# x축의 tick 위치를 설정 (set_ticks 호출)
ax1.set_xticks(range(len(years)))  # x축 위치 설정
ax1.set_xticklabels(years, rotation=45)  # x축 레이블 설정

# 두 번째 y축을 사용하여 1인당 GDP 그래프 추가
ax2 = ax1.twinx()
ax2.plot(years, gdp_values, marker='s', color='g', linestyle='--', label='1인당 GDP (USD)')
ax2.set_ylabel('1인당 GDP (USD)', color='g')  # 1인당 GDP의 y축 레이블을 오른쪽으로 이동
ax2.tick_params(axis='y', labelcolor='g')

# 타이틀 추가
plt.title('연도별 1인당 전력 소비량과 1인당 GDP 비교')

# 그래프 출력
fig.tight_layout()  # 레이아웃 조정
plt.grid(True)
plt.show()
```

연도별 1인당 전력 소비량과 1인당 GDP 비교(추세선 추가)

```python
import csv
import matplotlib.pyplot as plt
import numpy as np  # numpy 임포트 추가

# '1인당_국내총생산__당해년가격.csv' 파일을 열고 데이터 읽기
gdp_file = '1인당_국내총생산__당해년가격2.csv'
power_file = '연도별_전력수급_실적.csv'

# GDP 데이터 읽기
gdp_data = {}
f = open(gdp_file, encoding='cp949')  # 파일 열기
data = csv.reader(f)
header = next(data)  # 첫 번째 줄(헤더) 건너뛰기
for row in data:
    if row[0] == '대한민국':  # 대한민국 데이터만 사용
        for year, value in zip(header[2:], row[2:]):  # GDP 값이 세 번째 열부터 시작
            gdp_data[year] = float(value)
f.close()  # 파일 닫기

# 전력 소비량 데이터 읽기
power_data = {}
f = open(power_file, encoding='cp949')  # 파일 열기
data = csv.reader(f)
header = next(data)  # 첫 번째 줄(헤더) 건너뛰기
for row in data:
    year = row[0]
    if year in gdp_data:  # GDP 데이터에 있는 연도만 포함
        power_data[year] = float(row[-1])  # '인구1인당전력소비량 (kWh/년)' 열 사용
f.close()  # 파일 닫기

# 공통 연도 추출
years = sorted(set(gdp_data.keys()) & set(power_data.keys()))
gdp_values = [gdp_data[year] for year in years]
power_values = [power_data[year] for year in years]

# 시각화를 위한 두 개의 서브플롯 생성
fig, ax1 = plt.subplots(figsize=(10, 6))

# 첫 번째 그래프: 1인당 전력 소비량
ax1.plot(years, power_values, marker='o', color='b', label='1인당 전력 소비량 (kWh/년)')
ax1.set_xlabel('연도')
ax1.set_ylabel('1인당 전력 소비량 (kWh/년)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# x축의 tick 위치를 설정 (set_ticks 호출)
ax1.set_xticks(range(len(years)))  # x축 위치 설정
ax1.set_xticklabels(years, rotation=45)  # x축 레이블 설정

# 두 번째 y축을 사용하여 1인당 GDP 그래프 추가
ax2 = ax1.twinx()
ax2.plot(years, gdp_values, marker='s', color='g', linestyle='--', label='1인당 GDP (USD)')
ax2.set_ylabel('1인당 GDP (USD)', color='g')  # 1인당 GDP의 y축 레이블을 오른쪽으로 이동
ax2.tick_params(axis='y', labelcolor='g')

# 추세선 추가 (선형 회귀)
# 1인당 전력 소비량에 대한 추세선
power_slope, power_intercept = np.polyfit(range(len(years)), power_values, 1)  # 1차원 선형 회귀
power_trend = np.polyval([power_slope, power_intercept], range(len(years)))
ax1.plot(years, power_trend, color='purple', linestyle='-', linewidth=3, label='전력 소비량 추세선', marker='x', alpha=0.9)

# 1인당 GDP에 대한 추세선
gdp_slope, gdp_intercept = np.polyfit(range(len(years)), gdp_values, 1)  # 1차원 선형 회귀
gdp_trend = np.polyval([gdp_slope, gdp_intercept], range(len(years)))
ax2.plot(years, gdp_trend, color='orange', linestyle='-', linewidth=3, label='GDP 추세선', marker='x', alpha=0.5)

# 타이틀 추가
plt.title('연도별 1인당 전력 소비량과 1인당 GDP 비교')

# 범례 추가
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# 그래프 출력
fig.tight_layout()  # 레이아웃 조정
plt.grid(True)
plt.show()
```

Scatter 함수로 연도별 경제 성장률과 특정 분야의 산점도로 출력

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc

# 한글 폰트 설정
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

# CSV 파일 경로
file_path = "연도별_전력수급_실적.csv"

# 데이터 불러오기
try:
    data = pd.read_csv(file_path, encoding='cp949')  # 인코딩에 맞게 수정
    print("데이터 로드 성공!")
except Exception as e:
    print(f"데이터 로드 중 오류 발생: {e}")
    exit()

# X축 고정: 연도별 경제성장률
year_column = '시점'  # 연도
x_column = '경제성장률 (%)'  # X축: 경제성장률

# 데이터 확인
if year_column not in data.columns or x_column not in data.columns:
    print(f"데이터에 '{year_column}' 또는 '{x_column}' 열이 없습니다. 열 이름을 확인하세요.")
    print("사용 가능한 열 이름:", data.columns.tolist())
    exit()

# Y축 선택
print("\n[데이터 열 이름]")
print(data.columns.tolist())
print(f"\nX축은 '연도({year_column})'와 '{x_column}'로 고정됩니다.")

try:
    y_column = input("Y축에 사용할 열 이름을 입력하세요: ").strip()

    # 선택한 열 데이터 추출
    years = data[year_column]
    x_data = data[x_column]
    y_data = data[y_column]
except KeyError as e:
    print(f"잘못된 열 이름입니다: {e}")
    print("열 이름을 위 목록에서 확인한 후 다시 시도하세요.")
    exit()

# 산점도 출력
plt.figure(figsize=(12, 6))
plt.scatter(x_data, y_data, color='blue', alpha=0.7)

# 데이터 포인트에 연도 표시
for i, year in enumerate(years):
    plt.text(x_data[i], y_data[i], str(year), fontsize=8, alpha=0.7)

plt.title(f'{x_column}과 {y_column}의 산점도 (연도 표시)', fontsize=16)
plt.xlabel(x_column, fontsize=14)
plt.ylabel(y_column, fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```
